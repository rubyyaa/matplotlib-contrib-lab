```python
import numpy as np
import matplotlib.pyplot as plt

# Last 4 digits of Drexel ID used as the reproducible random seed.
seed = 1234  # TODO: replace with your own last 4 digits
rng = np.random.default_rng(seed)

n_readings = 200
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n_readings)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n_readings)
timestamps = rng.uniform(low=0.0, high=10.0, size=n_readings)

print(f"Generated {n_readings} readings per sensor.")
print(f"Sensor A mean/std (sample): {sensor_a.mean():.2f}/{sensor_a.std(ddof=1):.2f}")
print(f"Sensor B mean/std (sample): {sensor_b.mean():.2f}/{sensor_b.std(ddof=1):.2f}")
print(f"Timestamp range: {timestamps.min():.2f} to {timestamps.max():.2f} s")
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(timestamps, sensor_a, color='tab:blue', alpha=0.75, s=28, label='Sensor A')
plt.scatter(timestamps, sensor_b, color='tab:orange', alpha=0.75, s=28, label='Sensor B')

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor Temperature vs Time')
plt.legend()
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
plt.figure(figsize=(9, 5))
plt.hist(sensor_a, bins=30, alpha=0.5, color='tab:blue', label='Sensor A')
plt.hist(sensor_b, bins=30, alpha=0.5, color='tab:orange', label='Sensor B')

mean_a = sensor_a.mean()
mean_b = sensor_b.mean()
plt.axvline(mean_a, color='tab:blue', linestyle='--', linewidth=2, label=f'Sensor A mean: {mean_a:.2f} C')
plt.axvline(mean_b, color='tab:orange', linestyle='--', linewidth=2, label=f'Sensor B mean: {mean_b:.2f} C')

plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title('Temperature Distribution: Sensor A vs Sensor B')
plt.legend()
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
```

```python
overall_mean = np.concatenate([sensor_a, sensor_b]).mean()

plt.figure(figsize=(8, 5))
plt.boxplot([sensor_a, sensor_b], tick_labels=['Sensor A', 'Sensor B'])
plt.axhline(overall_mean, color='gray', linestyle='--', linewidth=2,
            label=f'Overall mean: {overall_mean:.2f} C')

plt.ylabel('Temperature (deg C)')
plt.title('Sensor Temperature Distribution Comparison')
plt.legend()
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```
